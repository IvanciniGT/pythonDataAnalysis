```python
#-1 means no idea of it at all
#incidents['StructuresDestroyed'].replace(np.NaN, -1)
incidents['StructuresDestroyed']=incidents['StructuresDestroyed'].fillna(-1).apply( lambda number : int(number) )
# incidents['StructuresDestroyed'].fillna(-1).astype( int )
```

```python
incidents['StructuresDestroyed'].unique()
```

```python
#Calculate the mean of this column?
#
# 10 structures were destroyed - small incidents
# 10-50 - medium incidents
# 50-500 - important incidents
# +500 - really important incidents

kind_of_fire = pd.cut( incidents['StructuresDestroyed'], (-2,-1,10, 50, 500, 999999999) , labels=['na', 'small', 'medium', 'important', 'really important'] )
kind_of_fire.value_counts()
```

```python
# we need to add that column to our dataframe
# groupby function 
# sum function
incidents['kind_of_fire']=kind_of_fire
```

```python
incidents.groupby('kind_of_fire').sum()['Helicopters']
incidents.groupby('kind_of_fire')['Helicopters'].sum() # Faster
```

```python
import re

def cleanUp(a_text):
    PATTERN = '((<[a-zA-Z0-9 /_=:".?-]*>)|[\n\r\t ])+'
    a_clean_text=re.sub( PATTERN , " ", str(a_text))
    return a_clean_text
```

```python
incidents['ConditionStatement'].apply( cleanUp ).unique()
incidents['ConditionStatement']=incidents['ConditionStatement'].apply( cleanUp )
incidents['ConditionStatement'].unique()
```
